```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re 
```

```python
wiki = requests.get("https://zh.wikipedia-on-ipfs.org/wiki/%E4%B8%AD%E5%9B%BD%E5%A4%A7%E9%99%86%E7%BD%91%E7%BB%9C%E7%94%A8%E8%AF%AD%E5%88%97%E8%A1%A8")

# scrape webpage
soup = BeautifulSoup(wiki.content, 'html.parser')

scrape_name = soup.find('h1').text.replace('\n',"")
scrape_file_name = scrape_name + '.csv'
```

```python
def get_slang(s):
    index = s.find("：")
    if index != -1:
        return s[:index]
    else:
        index = s.find(":")
        if index != -1:
            return s[:index]
        return s

```

```python
details = soup.find_all("details")

list = []
for i in details:
    list += i.find_all("li")

slang_list = []
for li in list:
    slang_list.append(get_slang(li.text))

slang_list_1 = []
for i in slang_list:
    #print(i)
    if i.find("/") != -1:
        temp = i.split('/')
        slang_list_1 += [re.sub("[\（\[].*?[\）\]]", "", j) for j in temp]
    elif i.find("／") != -1:
        temp = i.split('／')
        slang_list_1 += [re.sub("[\（\[].*?[\）\]]", "", j) for j in temp]
    else:
        slang_list_1.append(re.sub("[\（\[].*?[\）\]]", "", i))

slang_list_2 = [i for i in slang_list_1 if len(i) >= 2 and len(i) <= 6]

for i in slang_list_2:
    print(i)
```
